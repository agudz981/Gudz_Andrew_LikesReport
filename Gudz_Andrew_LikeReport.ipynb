<span style ="font-family:Georgia">Before seeing how many likes are recorded, its important to see the plain data. This is accomplished by opening the related json file titled "likes.json". The first task is to create a few variables on how the json file and the panda system will activate in the Python system. Pandas are a system of code that will greatly help sort the data later in the report.

```python
import pandas as pd
import json
```

<span style ="font-family:Georgia">Once those variables are setup correctly, two more variables are created to open the json. By establishing the variable "j" as the "open command" there is now a feasible way to open the file. And the "likes_dat" variable acts as a vessel so the data can finally be open.</span>

```python
with open ("/Users/andrewgudz/Data_Fall_2022/technogecko_20200714_toshare/likes.json") as j:
    likes_dat = json.load(j)
```

<span style ="font-family:Georgia">And eureka! The data is now viewable.</span>

```python
likes_dat
```

<span style ="font-family:Georgia">Can the data now tell the reader how many likes Dr. Silva has given? Well...not exactly. The data would be easier to sort and read if it was sorted in a table. This is accomplished by creating a panda "DataFrame". For practice, the "Timestamp" section of the data was put through a "DataFrame". The "Timestamp" section acts as a header for the soon to be table along with specific data collected from the previous variable "likes_dat". The "['media_likes']" further specifies the Timestamp data from this table.<span>

```python
likes_dat_df = pd.DataFrame({'Timestamp': likes_dat['media_likes']})
```

<span style ="font-family:Georgia">And once a new variable is created with the "DataFrame" command the table is now viewable.<span>

```python
likes_dat_df
```

<span style ="font-family:Georgia">Can a table based on how many likes now be created? Not yet. Its best to understand what categories of data are in the json. The first category, which was just created in a table, is the "Timestamps"; essential when the like happened. The second category is the "Accountnames"; the user name of each liked post. These two categories can be separated to better create the table.<span>

<span style ="font-family:Georgia">By creating two new variables, "timestamp" and "account_name", the data can be separated and used specifically for columns for the final table. Similar to a for loop in javascript the variable "x" can specify that the data should be taken from the previous table "likes_dat_df" and the specific data of ['Timestamp'].<span>

```python
timestamp = [x[0] for x in likes_dat_df['Timestamp']]
```

<span style ="font-family:Georgia">Now all the "Timestamps" can be displayed through a variable.</span>

```python
timestamp
```

<span style ="font-family:Georgia">Similar to the "timestamp" variable, an "account_name" variable is created and all that specific data can now be viewed.</span>

```python
account_name = [x[1] for x in likes_dat_df['Timestamp']]
```

```python
account_name
```

<span style ="font-family:Georgia">Now that the two variables are created, a categorized table can now be formed. "timestamp" and "account_name" variables can be used as columns for a new "DataFrame" chart.</span>

```python
table_df = pd.DataFrame({'Timestamps':timestamp,'Accountnames':account_name})
```

<span style ="font-family:Georgia">And now finally the data is now in a readable chart.</span>

```python
table_df
```

<span style ="font-family:Georgia">But how many likes were there total? There is no category of data called "likes" so how can the likes be viewed? The answer is revealed by counting the amount of "Timestamps". If "Accountnames" were counted there could be repeated names, meaning that some likes will not be counted. By using the "groupby" command and its further actions of "count()" and "sum()" the Timestamps can be counted. Through the amount of Timestamps there are a total of 335 likes.</span>

```python
table_df.groupby('Timestamps').count().sum()
```
